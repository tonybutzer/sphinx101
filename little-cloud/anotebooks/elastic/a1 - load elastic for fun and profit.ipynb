```python
# Import the os module, for the os.walk function
import os

song_list = []
# Set the directory you want to start from
rootDir = '/mnt/nuc/music-best'
for dirName, subdirList, fileList in os.walk(rootDir):
    #print('Found directory: %s' % dirName)
    for fname in fileList:
        if '.mp3' in fname:
            fullName = dirName + '/' + fname
            print(fullName)
            song_list.append(fullName)
            #print('%s\t%s' % (dirName,fname))
```

```python
from mp3_tagger import MP3File, VERSION_1, VERSION_2, VERSION_BOTH
```

```python
def index_line(id_number):
    index_str = '{ "index" : { "_index" : "music", "_type" : "song", "_id" : "%s" } }' % id_number
    return index_str
```

```python
file = open("music_success.log", "w")
file.close()

def logpath(path):
    file = open("music_success.log", "a")
    file.write(path)
    file.write('\n')
    file.close()
```

```python
import pprint
import json


def return_mp3(path_to_mp3):
    
    # Create MP3File instance.
    mp3 = MP3File(path_to_mp3)
    # Get all tags.
    try:
        tags = mp3.get_tags()
    except:
        print(path_to_mp3)
        return('{"none":"none"}')
    tagv2 = tags['ID3TagV2']
    #pp = pprint.PrettyPrinter(indent=4)

    if 'comment' in tagv2:
        del tagv2['comment']
    #pp.pprint(tagv2)
    tagv2['path'] = path_to_mp3
    logpath(path_to_mp3)
    song_json = json.dumps(tagv2)
    return(song_json)
```

```python
file = open("music_bulk.json", "w")
 
 
cnt = 0
for song in song_list:
    # pass
    # print(song)
    cnt = cnt + 1
    # if (cnt > 10):
        # break
    print(index_line(cnt))
    index_json = index_line(cnt)
    file.write(index_json)
    file.write("\n")

    song_json = return_mp3(song)
    file.write(song_json)
    file.write("\n")

file.write("\n")
file.close() 
```
